```python
# check duplicate row
df[df.duplicated(keep=False)]
```

```python
# check outlier
from scipy import stats
z_thr = 3.0
df[(np.abs(stats.zscore(df.select_dtypes(include=int))) > z_thr).any(axis=1)]
```

```python
# check missing data
df.isnull().sum()
```

```python
# check number of unique value
df.nunique()
```

```python
# check unique value
for feature in df.columns:
  print('------' + feature + '------')
  print(np.sort(df[feature].unique()))
  print()
```

DATA QUALITY REPORT:
* there are rows with repeating values but they are not identical
